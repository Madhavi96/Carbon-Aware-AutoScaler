```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# Assuming profiles is a list of directory paths containing the data_stats.csv files
# profiles = ['profile1', 'profile2', 'profile3']  # Example profiles

# Initialize a dictionary to store data for each service across profiles
service_data = {}

# Collect all unique service names from each profile
all_services = [set(pd.read_csv(os.path.join(profile, 'data_stats.csv'))['Name']) for profile in profiles]

# Find the intersection of services (services that exist in all profiles)
common_services = set.intersection(*all_services)

# Loop through each profile to collect data for common services only
for profile in profiles:
    data = pd.read_csv(os.path.join(profile, 'data_stats.csv'))
    
    # Filter data to include only common services
    data = data[data['Name'].isin(common_services)]
    
    # Store the median response time for each common service
    for index, row in data.iterrows():
        service_name = row['Name']
        median_response_time = row['Median Response Time']
        
        if service_name not in service_data:
            service_data[service_name] = []
        service_data[service_name].append((profile, median_response_time))

# Plotting
for service_name, response_times in service_data.items():
    profiles_list = [rt[0] for rt in response_times]
    median_times = [rt[1] for rt in response_times]
    
    plt.figure(figsize=(10, 5))
    plt.bar(profiles_list, median_times, color='skyblue')
    plt.xlabel('Profiles')
    plt.ylabel('Median Response Time')
    plt.title(f'Median Response Time for {service_name} across Profiles')
    plt.show()
```

```python
import requests

PROMETHEUS_URL = "http://localhost:9090/api/v1/query_range"
QUERY = "sum(kepler_container_joules_total{namespace='monitoring', container_namespace='default',mode='dynamic',container_name='ts-auth-service'})"

def fetch_energy_consumption(query, start_ts, end_ts, step='60s'):
    params = {
        "query": query,
        "start": start_ts,  # integer or float Unix timestamp
        "end": end_ts,      # integer or float Unix timestamp
        "step": step
    }

    response = requests.get(PROMETHEUS_URL, params=params)
    if response.status_code != 200:
        return f"Error: {response.status_code}, {response.text}"

    data = response.json()
    try:
        results = data["data"]["result"]
        if not results:
            return "No data returned for the specified query and time range."

        values = results[0]["values"]
        first_value = float(values[0][1])
        last_value = float(values[-1][1])
        energy_consumed = last_value - first_value
        return energy_consumed
    except (IndexError, KeyError, ValueError) as e:
        return f"Error parsing the result: {e}"

result = fetch_energy_consumption(QUERY, start_ts, end_ts, step="60s")
print("Energy consumed (Joules) between", start_ts, "and", end_ts, ":", result)
```
